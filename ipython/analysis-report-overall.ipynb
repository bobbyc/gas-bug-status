```python
import pandas as pd
import numpy as np
import json 
from pprint import pprint
import matplotlib.pyplot as plt

firefox = ['../data/Firefox50_bugs.json', '../data/Firefox51_bugs.json', '../data/Firefox52_bugs.json',
           '../data/Firefox53_bugs.json', '../data/Firefox54_bugs.json', '../data/Firefox55_bugs.json']

core = ['../data/Core50_bugs.json', '../data/Core51_bugs.json', '../data/Core52_bugs.json',
           '../data/Core53_bugs.json', '../data/Core54_bugs.json', '../data/Core55_bugs.json']

android = ['../data/Android50_bugs.json', '../data/Android51_bugs.json', '../data/Android52_bugs.json',
           '../data/Android53_bugs.json', '../data/Android54_bugs.json', '../data/Android55_bugs.json']

df = pd.DataFrame()
for target in firefox:
    with open(target) as data_file:
        data = json.load(data_file)
        dtemp = pd.DataFrame(data)
        dtemp = pd.DataFrame([dtemp['bugs'][n] for n in range(0, len(dtemp.index))])
        df = pd.concat([df, dtemp])
    
dfirefox = df
```

```python
df = pd.DataFrame()
for target in core:
    with open(target) as data_file:
        data = json.load(data_file)
        dtemp = pd.DataFrame(data)
        dtemp = pd.DataFrame([dtemp['bugs'][n] for n in range(0, len(dtemp.index))])
        df = pd.concat([df, dtemp])
    
dcore = df
```
